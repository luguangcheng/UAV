```python
import pandas as pd
import numpy as np
import math
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from scipy.stats import pearsonr
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import RandomizedLasso
from sklearn.feature_selection import RFE
```

```python
train_data = pd.read_csv('train_data_1130.csv',header=0) #多段划分 使用100m的数据训练
X = train_data.drop('label',axis=1)
y = train_data['label']
print X.shape,y.shape
```

```python
s1 = SelectKBest(chi2).fit(X,y).scores_
s1[np.isnan(s1)]=0
s1=s1.reshape(-1,1)
s1=MinMaxScaler().fit_transform(s1)
s1=map(lambda x:round(x,4),s1)
s1=np.array(s1)
s1
#MinMaxScaler().fit_transform(s1)
```

```python
s1 = SelectKBest(chi2).fit(X,y).scores_
s1[np.isnan(s1)]=0
s1=s1.reshape(-1,1)
s1
```

```python
# 皮尔森系数
s2 = X.apply(lambda x:abs(pearsonr(x,y)[0])).values
s2[np.isnan(s2)]=0
s2=MinMaxScaler().fit_transform(s2)
s2=map(lambda x:round(x,4),s2)
s2=np.array(s2)
s2
```

```python
s6 = SelectFromModel(LogisticRegression(penalty="l2", C=1)).fit(X, y)
l2 = s6.estimator_
s3=abs(l2.coef_[0])
s3=MinMaxScaler().fit_transform(s3)
s3=map(lambda x:round(x,4),s3)
s3=np.array(s3)
s3
```

```python
s = SelectFromModel(LogisticRegression(penalty="l1", C=1)).fit(X, y)
l1 = s.estimator_
s4=abs(l1.coef_[0])
s4=MinMaxScaler().fit_transform(s4)
s4=map(lambda x:round(x,4),s4)
s4=np.array(s4)
s4
```

```python
s =SelectFromModel(GradientBoostingClassifier()).fit(X, y)
gb = s.estimator_
s5=gb.feature_importances_
s5=MinMaxScaler().fit_transform(s5)
s5=map(lambda x:round(x,4),s5)
s5=np.array(s5)
s5
```

```python
s =SelectFromModel(RandomForestRegressor()).fit(X, y)
rf = s.estimator_
s6=rf.feature_importances_
s6=MinMaxScaler().fit_transform(s6)
s6=map(lambda x:round(x,4),s6)
s6=np.array(s6)
s6
```

```python
rlasso = RandomizedLasso()  # 默认通过AIC选择正则化系数alpha
rlasso.fit(X, y)
s7=rlasso.scores_
s7=MinMaxScaler().fit_transform(s7)
s7=map(lambda x:round(x,4),s7)
s7=np.array(s7)
s7
```

```python
lr = LogisticRegression(penalty='l1')
#rank all features, i.e continue the elimination until the last one
rfe = RFE(lr, n_features_to_select=1)
rfe.fit(X,y)
s8=rfe.ranking_
n=len(s8)
s8=map(lambda x:round((n+1-x)/float(n),4),s8)
s8=np.array(s8)
s8
```

```python
s=s1+s2+s3+s4+s5+s6+s7+s8
s
```

```python
s_sum = pd.Series(s,index=X.columns)
s_sum.sort_values(ascending=False,inplace=True)
s_sum
```

```python
# 递归特征消除（REF）
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

names = X.columns
#use linear regression as the model
lr = LogisticRegression(penalty='l2')
#rank all features, i.e continue the elimination until the last one
rfe = RFE(lr, n_features_to_select=1)
rfe.fit(X,y)

print "Features sorted by their rank:"
print sorted(zip(map(lambda x: round(x, 4), rfe.ranking_), names))
```
